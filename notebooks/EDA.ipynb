### 3.3 Limpieza y normalización
- Corrección de tipos. +
- Tratamiento de nulos. +
- Eliminación o tratamiento de duplicados. +
- Normalización de categorías y fechas. +
- Justificación de cada decisión. +

```python
#Vamos a eliminar los duplicados y comprobamos que efectivamente se han eliminado
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df
```

Vemos como el Dataset ha pasado de 371528 filas a 371524 filas, ya que hemos eliminado las 4 que habia duplicadas.

```python
#Para cambiar las columnas de fechas que son de tipo objeto a tipo fechas
fecha_columnas = ['dateCrawled', 'dateCreated', 'lastSeen']
for columna in fecha_columnas:
    df[columna] = pd.to_datetime(df[columna])

df.dtypes
```

Una vez ejecutado el codigo anterior, comprobamos que los tipos de datos han cambiado en las columnas.

```python
#Encotramos las columnas que tienen nulos
columnas_con_nulos = df.columns[df.isnull().any()]
print(columnas_con_nulos)
```

```python
for columna in columnas_con_nulos:
    df[columna] = df[columna].fillna('Desconocido')
    
df.info
```

Ahora, en los valores donde aparecia 'NaN' ahora aprece 'Desconocido'. Esto lo hacemos para no borrar muchos datos porque como habiamos visto antes, habia muchas filas con valores 'NaN'.

```python
#Para manejar mejor los datos, vamos a pasar toda la informacion de tipo objeto a minusculas
#y vamos a traducir las palabras que aparecen en aleman al español
str_cols = df.select_dtypes(include=['object']).columns
for col in str_cols:
    df[col] = df[col].str.lower()

mapping_gearbox = {'manuell': 'manual', 'automatik': 'automatico'}
mapping_damage = {'nein': 'no', 'ja': 'si'}
mapping_fuel = {
    'benzin': 'gasolina', 
    'diesel': 'diesel', 
    'lpg': 'glp', 
    'cng': 'gnc', 
    'hybrid': 'hibrido', 
    'elektro': 'electrico'
}

df['gearbox'] = df['gearbox'].replace(mapping_gearbox)
df['notRepairedDamage'] = df['notRepairedDamage'].replace(mapping_damage)
df['fuelType'] = df['fuelType'].replace(mapping_fuel)

df.head(20)
```

Como se puede ver, ahora toda la información aparece en letras minisculas y hemos traducido con exito los valores especificados en los diccionarios.

Por qué hemos hecho cada cambio:
- Hemos eliminado las filas duplicadas para que no se nos vicien los datos. Aunque en nuestro caso eran 4 filas, en otros Datasets puede darse el caso de que sean cientas o miles.
- Hemos cambiado las columnas con fechas de tipo objeto a tipo date para poder extraer y filtrar por mes/año o rango de fechas de una manera más sencilla.
- Hemos tratado los datos 'Nan' como 'Desconocidos' porque si hubieramos eliminado las filas en las que habia valores nulos se nos habría reducido el Dataset de manera considerable, haciendo que ya no fuera lo mas 'realista' posible y pudiendo quedar inservible.
- Hemos pasado a minusculas toda la información para que no se vieran commo categorias diferentes información que sabemos de primera mano que quiere decir lo mismo, por ejemplo: 'Suv' y 'suv'.
- Hemos traducido del aleman aquellas palabras que nos podrian dificultar mas la comprensión de las tablas para que nos cueste menos el tratamiento de los datos.

*Cabe resaltar que cada cambio efectuado se ha comprobado y se ha producido correctamente.

### 3.4 Visualizaciones básicas
Debe incluir, como mínimo:
- Un histograma. +
- Una gráfica de barras. +
- Una visualización adicional que aporte información relevante. +

```python
sns.set(style="whitegrid")

# Eliminao precios y años poco realistas 
df_clean = df[
    (df['price'].between(500, 70000)) & 
    (df['yearOfRegistration'].between(1990, 2016))
].copy()

# Creo figura para los gráficos
plt.figure(figsize=(20, 15))

# 1. Distribución de Precios
plt.subplot(3, 2, 1)
sns.histplot(df_clean['price'], bins=50, kde=True, color='skyblue')
plt.title('Precios (500€ - 70.000€)')
plt.xlabel('Precio (€)')
plt.ylabel('Frecuencia')
```

```python
# Creo figura para los gráficos
plt.figure(figsize=(20, 15))

# 2. Año de Registro
plt.subplot(3, 2, 2)
sns.histplot(df_clean['yearOfRegistration'], bins=26, kde=False, color='salmon')
plt.title('Años de Registro (1990 - 2016)')
plt.xlabel('Año de Registro')
plt.ylabel('Cantidad de Autos')
```

```python
# Creo figura para los gráficos
plt.figure(figsize=(20, 15))

# 3. Top 10 Marcas más comunes
plt.subplot(3, 2, 3)
top_brands = df_clean['brand'].value_counts().head(10)

sns.barplot(x=top_brands.values, y=top_brands.index, palette='viridis')
plt.title('Top 10 Marcas')
plt.xlabel('Cantidad de Autos')
```

```python
# Creo figura para los gráficos
plt.figure(figsize=(20, 15))

# 4. Tipos de Vehículo
plt.subplot(3, 2, 4)
type_counts = df_clean['vehicleType'].value_counts()

sns.barplot(x=type_counts.index, y=type_counts.values, palette='magma')
plt.title('Tipo de Vehículo')
plt.xticks(rotation=45)
plt.ylabel('Cantidad')
```

```python
# Creo figura para los gráficos
plt.figure(figsize=(20, 15))

# 5. Boxplot de Precio vs Tipo de Vehículo
plt.subplot(3, 1, 3)

sns.boxplot(x='vehicleType', y='price', data=df_clean, palette='coolwarm')
plt.title('Precios por Tipo de Vehículo')
plt.xlabel('Tipo de Vehículo')
plt.ylabel('Precio (€)')
```
